```python
import dpkt
import datetime
import socket

arr = ["1. Google.pcap", "2. example.pcap", "3. httpforever.pcap", "4. FTP.pcap", "5. SSH.pcap"]

destination_ips = set()
dest_timestamp = {}

for i in arr:
    print("*" * 100)
    print("File Name: ", i)
    f = open(i, 'rb')

    pcap = dpkt.pcap.Reader(f)

    protocol_count = {'DNS': 0, 'HTTP': 0, 'HTTPS': 0, 'FTP': 0}


    for timestamp, buf in pcap:
        eth = dpkt.ethernet.Ethernet(buf)
        ip = eth.data
        
        if isinstance(ip, dpkt.ip.IP):
            time = datetime.datetime.utcfromtimestamp(timestamp)
            protocol = ip.p
            dest_ip = socket.inet_ntop(socket.AF_INET, ip.dst)
            
            destination_ips.add(dest_ip)
            dest_timestamp[dest_ip] = time
            
            if protocol == dpkt.ip.IP_PROTO_UDP:
                udp = ip.data

                if udp.dport == 53 or udp.sport == 53:
                    time = datetime.datetime.utcfromtimestamp(timestamp)
                    print("Time:", time)
                    print("Destination IP:", dest_ip)
                    print("Protocol: DNS")
                    protocol_count['DNS'] += 1
                    print()
                    
            elif protocol == dpkt.ip.IP_PROTO_TCP:
                tcp = ip.data

                if tcp.dport == 80 or tcp.sport == 80:
                    time = datetime.datetime.utcfromtimestamp(timestamp)
                    print("Time:", time)
                    print("Destination IP:", dest_ip)
                    print("Protocol: HTTP")
                    protocol_count['HTTP'] += 1
                    print()
                    
                elif tcp.dport == 443 or tcp.sport == 443:
                    time = datetime.datetime.utcfromtimestamp(timestamp)
                    print("Time:", time)
                    print("Destination IP:", dest_ip)
                    print("Protocol: HTTPS")
                    protocol_count['HTTPS'] += 1
                    print()
                    
                elif tcp.dport == 20 or tcp.sport == 20 or tcp.dport == 21 or tcp.sport == 21:
                    time = datetime.datetime.utcfromtimestamp(timestamp)
                    print("Time:", time)
                    print("Destination IP:", dest_ip)
                    print("Protocol: FTP")
                    protocol_count['FTP'] += 1
                    print()

                
                    

    print("Destination IP addresses and their timestamp:")
    for ip in destination_ips:
        print(f"Destination IP: {ip:20} Timestamp: {dest_timestamp[ip]}")

    
    print("Protocol counts for {}: ".format(i))
    for protocol, count in protocol_count.items():
        print(f"{protocol}: {count}")


    f.close()
```
